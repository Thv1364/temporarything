```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['9th Arial', '9th TNR', '10th Arial', '10th TNR', '11th Arial', '11th TNR', '12th Arial', '12th TNR']
values = [6.5, 7.2, 7.5, 7.5, 7.6666666667, 6.6, 7, 6.83333333333]
#bar_colors = ['tab:red', 'tab:blue','tab:red', 'tab:blue','tab:red', 'tab:blue','tab:red', 'tab:blue','tab:red', 'tab:blue']
bar_colors=[(0.8,0.8,0.8), (0.2,0.2,0.2)]*((len(categories)//2))
# bar_colors.append('magenta')
# bar_colors.append('yellow')

# Create the bar graph
plt.bar(categories, values,color=bar_colors,edgecolor='black', linewidth=1.6)
#colors = ['skyblue', 'lightcoral'] * (len(categories) // 2) 


# Add labels and title
plt.xlabel('Grades/Font')
plt.ylabel('Amount of Grammar/Spelling Mistakes (out of 11)')
plt.title('Arial vs. Times New Roman for Each Grade')
plt.xticks(rotation=45, ha='right') # Rotate labels by 45 degrees and align to the right
plt.tight_layout()
# Display the plot
plt.savefig('figure2.png', dpi=3000)
plt.show()
```
